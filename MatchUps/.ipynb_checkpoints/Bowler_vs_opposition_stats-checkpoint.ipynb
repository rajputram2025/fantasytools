```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np 
import os
import pandas as pd
```

```python
df1 = pd.read_csv('../archive/IPL_Ball_by_Ball_2008_2022.csv')
df1.dataframeName = 'deliveries.csv'
```

```python
df1.head(5)
```

```python
player_subset = df1[df1["bowler"] == 'Mohammed Shami']
player_subset
```

```python
df2 = pd.read_excel('../archive/batter_team_data(2022).xlsx')
df2.dataframeName = 'batter.csv'
df2
```

```python
batter_subset = df2[df2["Team"] == 'Royal Challengers Bangalore']
batter_subset = batter_subset["Batters"].to_numpy()
batter_subset
```

```python
for batter in batter_subset:

    print("Shami vs", batter)
    
    temp = player_subset[player_subset["batter"] == batter]
    
    if len(temp)==0:
        print("Not yet faced")
        
    else:

        extras_wides = temp[temp["extra_type"]=='wides'] 
        extras_noballs =  temp[temp["extra_type"]=='noballs']
        
        total_innings = len(pd.unique(temp['ID']))
        print(total_innings)

        total_runs = temp["batsman_run"].sum() + extras_wides["total_run"].sum() + extras_noballs["total_run"].sum()

        total_outs = temp["player_out"].isnull().sum()
        total_outs = len(temp)-total_outs
        print(total_outs)

        if total_outs == 0:
            print("Wicket not taken")
            
        else:
            average = total_runs/total_outs
            print(average)
            
        if total_outs == 0:
            print("Wicket not taken")
        else:
            strike_rate = len(temp)/ total_outs
            print(strike_rate)

        economy = total_runs / (len(temp)/6)
        print(economy)
```
