```python
df2 = pd.read_excel('../archive/bowler_team_data(2022).xlsx')
df2.dataframeName = 'bowler.csv'
df2
```

```python
bowler_subset = df2[df2["Team"] == 'Royal Challengers Bangalore']
bowler_subset = bowler_subset["Bowlers"].to_numpy()
bowler_subset
```

```python
player_subset = df1[df1["batter"] == 'DA Warner']

for bowler in bowler_subset:

    print("Warner vs", bowler)
    
    temp = player_subset[player_subset["bowler"] == bowler]
    
    if len(temp)==0:
        print("Not yet faced")
        
    else:
        total_runs = temp["batsman_run"].sum()
        print(total_runs)

        total_innings = len(pd.unique(temp['ID']))
        print(total_innings)

        total_outs = temp[temp["player_out"] == 'DA Warner']
        print(len(total_outs))
        
        dots = temp[temp["batsman_run"] == 0]
        dot_percentage = len(dots)/len(temp) * 100
        print(dot_percentage)

        boundaries = temp[temp["batsman_run"] == 4]
        sixes =  temp[temp["batsman_run"] == 6]
        boundary_percentage = len(boundaries) + len(sixes) / len(temp) * 100
        print(boundary_percentage)

        if len(total_outs)==0:
            print("Avg : inf")
        else:
            average = total_runs / len(total_outs)
            print(average)

        strike_rate = total_runs / len(temp) * 100
        print(strike_rate)
```
