```python
player_subset = df1[df1["batter"] == 'SN Khan']
player_subset
```

```python
df2 = pd.read_csv('../archive/IPL_Matches_2008_2022.csv')
df2.dataframeName = 'matches.csv'
```

```python
df2.head(5)
```

```python
team_subset1 = df2[df2["Team2"] == 'Royal Challengers Bangalore'] 
team_subset2 = df2[df2["Team1"] == 'Royal Challengers Bangalore'] 

match_id_list_1 = team_subset1["ID"].to_numpy()
match_id_list_2 = team_subset2["ID"].to_numpy()

match_id_list = np.concatenate([match_id_list_1,match_id_list_2])
print(match_id_list)
```

```python
m = len(match_id_list)
```

```python
player_subset = player_subset[player_subset['BattingTeam']!='Royal Challengers Bangalore']
```

```python
final_df = []
for i in range(len(player_subset)):
    if player_subset.iat[i,0] in match_id_list:
        final_df.append(player_subset.iloc[i])
```

```python
final_df
```

```python
final_df = pd.DataFrame(final_df)
```

```python
final_df
```

```python
#Overall
if len(final_df)!=0:
    
    total_runs = final_df["batsman_run"].sum()
    print(total_runs)

    total_innings = len(pd.unique(final_df['ID']))
    print(total_innings)

    total_outs = final_df[final_df["player_out"] == 'DA Warner']
    print(len(total_outs))

    if len(total_outs)==0:
        print("Player not yet dismissed at ground")
    
    else:
        average = total_runs / len(total_outs)
        print(average)

    strike_rate = total_runs / len(final_df) * 100
    print(strike_rate)

    dots = final_df[final_df["batsman_run"] == 0]
    dot_percentage = len(dots)/len(final_df) * 100
    print(dot_percentage)

    boundaries = final_df[final_df["batsman_run"] == 4]
    sixes =  final_df[final_df["batsman_run"] == 6]
    boundary_percentage = len(boundaries) + len(sixes) / len(final_df) * 100
    print(boundary_percentage)

else:
    print("Player not played at ground")
```

```python
#1st_innings
if len(final_df)==0:
    print("Player not played in ground")
    
else:
    final_df_1 = final_df[final_df["innings"]==1]

    if len(final_df_1)!=0:

        total_runs = final_df_1["batsman_run"].sum()
        print(total_runs)

        total_innings = len(pd.unique(final_df_1['ID']))
        print(total_innings)

        total_outs = final_df_1[final_df_1["player_out"] == 'DA Warner']
        print(len(total_outs))

        if len(total_outs)==0:
            print("Player not yet dismissed at ground in 1st innings")

        else:
            average = total_runs / len(total_outs)
            print(average)

        strike_rate = total_runs / len(final_df_1) * 100
        print(strike_rate)

        dots = final_df_1[final_df_1["batsman_run"] == 0]
        dot_percentage = len(dots)/len(final_df_1) * 100
        print(dot_percentage)

        boundaries = final_df_1[final_df_1["batsman_run"] == 4]
        sixes =  final_df_1[final_df_1["batsman_run"] == 6]
        boundary_percentage = len(boundaries) + len(sixes) / len(final_df_1) * 100
        print(boundary_percentage)
    else:
            print("Player not played at ground in 1st innings")
```

```python
#2nd_innings
if len(final_df)==0:
    print("Player not played in ground")
    
else:
    final_df_2 = final_df[final_df["innings"]==2]
    
    if len(final_df_2)!=0:
        total_runs = final_df_2["batsman_run"].sum()
        print(total_runs)

        total_innings = len(pd.unique(final_df_2['ID']))
        print(total_innings)

        total_outs = final_df_2[final_df_2["player_out"] == 'DA Warner']
        print(len(total_outs))

        if len(total_outs)==0:
                print("Player not yet dismissed at ground in 2nd innings")

        else:
            average = total_runs / len(total_outs)
            print(average)

        strike_rate = total_runs / len(final_df_2) * 100
        print(strike_rate)

        dots = final_df_2[final_df_2["batsman_run"] == 0]
        dot_percentage = len(dots)/len(final_df_2) * 100
        print(dot_percentage)

        boundaries = final_df_2[final_df_2["batsman_run"] == 4]
        sixes =  final_df_2[final_df_2["batsman_run"] == 6]
        boundary_percentage = len(boundaries) + len(sixes) / len(final_df_2) * 100
        print(boundary_percentage)
        
    else:
            print("Player not played at ground in 2nd innings")
```
